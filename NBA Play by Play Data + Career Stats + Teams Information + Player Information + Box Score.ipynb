# Box Score for a given game or for all of the season

```python
from nba_api.stats.endpoints import boxscoretraditionalv2

player_stat_data=boxscoretraditionalv2.BoxScoreTraditionalV2(game_id=game_id)
stats_df=player_stat_data.get_data_frames()[0]
stats_df.tail()
```

```python
#function that gets box score data based on game_ids
def get_box_score_data(game_id):
    player_stat_data = boxscoretraditionalv2.BoxScoreTraditionalV2(game_id=game_id, headers=headers, timeout=100000000)
    df=player_stat_data.player_stats.get_data_frame()
    # df=player_stat_data.team_stats.get_data_frame() if we want the stats for the team
    return df
```

Obtener los game_id de cualquier temporada

```python
from nba_api.stats.static import teams
from nba_api.stats.endpoints import LeagueGameFinder

# Obtener los IDs de los equipos NBA
nba_teams = teams.get_teams()

# Obtener los game_id de la temporada que quiera usando LeagueGameFinder
gamefinder = LeagueGameFinder(season_nullable='2022-23')
games = gamefinder.get_data_frames()[0]

# Obtener los game_id de los partidos
game_ids = games['GAME_ID'].tolist()
```

```python
#loop for a dataframe with all of the boxscore of all players in one season
boxscores=[]

for game_id in game_ids:
    boxscore_data=get_box_score_data(game_id)
    boxscores.append(boxscore_data)
    
final_df=pd.concat(boxscores, ignore_index=True)
```
